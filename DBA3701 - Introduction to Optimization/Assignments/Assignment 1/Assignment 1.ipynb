```python
# Q2
A = np.array([[2, 0],
              [1, 0]])
B = np.array([[0, 0], 
              [3, 1]])

print(np.matmul(A, B))
```

```python
# Q3
A = np.array([[1, 1],
              [-1, 1]])
B = np.array([[0.5, -0.5], 
              [0.5, 0.5]])

print(np.matmul(A, B))
print(np.matmul(B, A))
```

```python
# Q4a
A = np.array([[1, -2, 3],
              [4, -5, 7]])
B = np.array([[1, 2], 
              [1, 1],
              [2, 0]])

print(np.matmul(A, B))
print(np.matmul(B, A))

# Q4b
print(np.matmul(A, B).T)
print(B.T)
print(A.T)
print(np.matmul(B.T, A.T))
```

```python
# Q5
A = np.array([[2, 1],
              [-1, 0]])
A_inv = np.linalg.inv(A)

print(A_inv)
print(A.T)
print(np.linalg.inv(A.T))
print(A_inv.T)
```

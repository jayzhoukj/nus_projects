```python
### Import required packages and set cwd
import numpy as np, pandas as pd, re, os
os.chdir(r"C:\Users\Kai Jing\Desktop\NUS\Business (Accountancy)\NUS BAC\Sem 3.2\DBA3803 - Predictive Analytics in Business\Assignment 1\Datasets")

### Part 1 - Read the first sheet, specifying headers
df = pd.read_excel("TBill-Dows-CPI-STI-UKBirthsDeaths-BitCoin.xlsx",
                   header=0,
                   index_col=0,
                   sheet_name=0)

r = re.compile('^Unnamed: ')
empty_columns = len(list(filter(r.match, list(df.columns))))/len(list(df.columns))
target_empty_columns = 0.2
skip_rows = 0

## Keep skipping rows on the top of the sheet if the there are a lot of empty column names
while empty_columns >= target_empty_columns:
    skip_rows += 1
    
    df = pd.read_excel("TBill-Dows-CPI-STI-UKBirthsDeaths-BitCoin.xlsx",
                       skiprows=skip_rows,
                       header=0,
                       index_col=0,
                       sheet_name=0)
    
    empty_columns = len(list(filter(r.match, list(df.columns))))/len(list(df.columns))

df.head()
```

```python
### Part 2 - Specify target variable
target_variable = 'Bitcoinity Bitcoin Price'
```

```python
### Part 3 - Use df.drop() to remove all columns with names containing 'Bitcoin', except for the target variable
# print(list(df.columns))

for column_name in list(df.columns):
    if 'Bitcoin' in column_name:
        if column_name == target_variable:
            continue
        else:
            df.drop(column_name, axis=1, inplace=True)

print(list(df.columns))
```

```python
### Part 4 - Remove the 2 columns 'Consumer Price Index-Urban, CPI-U' and 'Inflation'
remove_columns = ['Consumer Price Index-Urban, CPI-U', 'Inflation']

df.drop(columns=remove_columns,
        axis=1,
        inplace=True)

print(list(df.columns))
```

```python
### Part 5 - Use 'import re', then df.rename() and re.sub('\W',) to change all non-alphanumeric characters 
### in the variable names to "_"
import re

rename_columns = {}

for column_name in list(df.columns):
    ## Go through each column name and check if it contains non-alphanumeric
    if re.search('\W', column_name):
        ## Remove non-alphanumeric at the start or end of the column names or replace with '_' if in the middle
        new_name = re.sub('^\W+', '', column_name)
        new_name = re.sub('\W+$', '', new_name)
        rename_columns.update({column_name: re.sub('\W+', '_', new_name)})

df.rename(columns=rename_columns,
          inplace=True)

print(list(df.columns))

## Update the target variable formatting (remove non-alphanumeric at start & end of string, else replace with '_')
target_variable = re.sub('^\W+', '', target_variable)
target_variable = re.sub('\W+$', '', target_variable)
target_variable = re.sub('\W+', '_', target_variable)
```

```python
### Part 6 - Any name beginning with a digit (i.e., a numeral) needs to be prepended by '_'
prepend_digits = {}

for column_name in list(df.columns):
    if re.match('\d', column_name[0]):
        new_name = '_'+column_name
        prepend_digits.update({column_name: new_name})

df.rename(columns=prepend_digits,
          inplace=True)

print(list(df.columns))

## Update the target variable formatting (prepend if first char is a digit)
if re.match('\d', target_variable[0]):
    target_variable = '_'+target_variable
```

```python
### Part 7 - Delete all rows with any missing value, via df.dropna()
df.dropna(axis=0,
          how='any',
          inplace=True)

df.reset_index(drop=True,
               inplace=True)

df.head()
```

```python
### Part 8 - Create an empty dataframe using pd.DataFrame() with no argument
df_corr = pd.DataFrame()
```

```python
### Part 9 - Use loc[x, y] on the dataframe created above to contain the correlation of x & y, 
### for all Xs' correlations with the target variable
# df_corr = abs(df.corr()).loc[:,target_variable].to_frame()

df_corr = df_corr.append(df.corr())

df_corr = df_corr.loc[:,target_variable].to_frame()

df_corr.head()
```

```python
### Part 10 - Sort your dataframe by the absolute values of the correlations, in descending ordere, and print it
df_corr.drop(target_variable, 
             axis=0,
             inplace=True)

df_corr = abs(df_corr)

df_corr.sort_values(by=[target_variable],
                    ascending=False,
                    inplace=True)

print(df_corr)
```

```python
### Part 11 - Create a list of the variable names using list(df), then move y's name to the front
col_names = list(df)
col_names.insert(0, col_names.pop(col_names.index(target_variable)))

print(col_names)
```

```python
### Part 12 - Use the list above to make the model equation for statsmodel ols
model_equation = ' + '.join(col_names).replace('+', '~', 1)

print(model_equation)
```

```python
### Part 13 - Output the summary for the regression of Y on all Xs using ols
from statsmodels.formula.api import ols
regression_output = ols(model_equation, df).fit()

regression_output.summary2()
```

```python
### Part 14 - Use dir() to find out how to extract Xs' t-statistics from the regression output 
### into a single-column dataframe
print(dir(regression_output))

df_tvalues = regression_output.tvalues.rename('t-statistics').to_frame()

df_tvalues.head()
```

```python
### Part 15 - Sort your dataframe by the absolute values of the t-statistics, in descending order, and print it
df_tvalues.drop('Intercept',
                axis=0,
                inplace=True)

df_tvalues = abs(df_tvalues)

df_tvalues.sort_values(by=['t-statistics'],
                       ascending=False,
                       inplace=True)

print(df_tvalues)
```

```python
### Part 16 - List the names of the sorted variables for correlation (Step 10) and the names of the sorted variables 
### for regression (Step 15), side by side
df_variables = pd.DataFrame(data={'Correlation': list(df_corr.index), 'Regression': list(df_tvalues.index)})

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_variables)
```

```python
### Part 17 - Use '#' to write a short paragraph on what you discover about the relationship between correlation 
### and regression in Step 16

# By comparing the independent variables from the correlation and regression results, we observe that when variables 
# are ranked on top of the sorted list of variables for correlation (such as the 4 DJIA indicators), they tend to 
# be ranked lower on the sorted list for regression (DJIA variables fall in the range of 10th to 14th variable).
# The same trend can be seen for most other variables such as the 52, 26, 13 and 4 week coupon equivalents and 
# bank discounts, except for ST_Index_Low and Weekday_Moday_1. Hence, we can say that there is an observation of
# a weak negative correlation between the sorted variables when comparing the absolute strength of their correlation
# with the target variable and the absolute values of their t-value statistics.
```
